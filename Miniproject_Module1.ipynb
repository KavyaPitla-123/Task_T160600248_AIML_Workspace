```python
top_subscribers = yt[['Youtuber', 'subscribers']]\
                    .sort_values(by='subscribers', ascending=False)

print(top_subscribers.head(10))
```

```python
top_views = yt[['Youtuber', 'video views']]\
                .sort_values(by='video views', ascending=False)

print(top_views.head(10))
```

```python
top_subscribers = yt.nlargest(10, 'subscribers')

plt.figure(figsize=(10,6))
sns.barplot(x='subscribers', y='Youtuber', data=top_subscribers)

plt.title('Top 10 YouTube Channels by Subscribers')
plt.xlabel('Subscribers')
plt.ylabel('YouTube Channel')

plt.show()
```

```python
top_views= yt.nlargest(10,'video views')
plt.figure(figsize=(10,6))
sns.barplot(x='video views',y='Youtuber',data=top_views)
plt.title('Top 10 YouTube Channels by Video Views')
plt.xlabel('Video Views')
plt.ylabel('YouTube Channel')
plt.show()
```

```python
country_counts = yt['Country'].value_counts()

print(country_counts.head())
```

```python
top_countries = yt['Country'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.barh(top_countries.index, top_countries.values)

plt.title('Top 10 Countries by Number of YouTube Creators')
plt.xlabel('Number of Creators')
plt.ylabel('Country')

plt.show()
```
