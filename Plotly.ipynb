```python
import pandas as pd
import plotly.express as px
df = pd.DataFrame({
    "Govt_Service":[6,7,9,10,11],
    "Salary":[50000,60000,80000,90000,100000]
})
df
```

```python
fig = px.bar(df,x="Govt_Service",y="Salary",color="Salary",title="Govt_Service and Salary")
fig.show()
```

```python
fig = px.histogram(df,x="Govt_Service",y="Salary",color="Salary",title="Govt_Service and Salary")
fig.show()
```

```python
fig = px.pie(df, names="Govt_Service", values="Salary", color="Salary", title="Govt_Service and Salary")
fig.show()
```

```python
fig = px.scatter(df,x="Govt_Service",y="Salary",color="Salary",title="Govt_Service and Salary")
fig.show()
```

```python
fig = px.line(df,x="Govt_Service",y="Salary",title="Govt_Service and Salary")
fig.show()
```
