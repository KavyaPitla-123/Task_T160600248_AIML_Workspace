```python
from google.colab import files
uploaded = files.upload()
file_name=list(uploaded.keys())[0]
print("uploaded file:",file_name)
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(file_name)
df.head()
```

```python
df.info()
df.describe()
df.isnull()
df.isnull().sum()
```

```python
df['date'] = df['date'].astype(str).str.strip()
df['date'] = pd.to_datetime(df['date'],errors='coerce')
df.info()
df.head()
```

```python
nums_cols=["rainfall","temperature","humidity","wind_speed"]
df[nums_cols]=df[nums_cols].fillna(df[nums_cols].mean())
df["weather_condition"] = df["weather_condition"].fillna(df["weather_condition"].mode()[0])
print(df.isnull().sum())
```

```python
df['weather_condition']=df['weather_condition'].str.lower()

df
```

```python
df_no_dup = df.drop_duplicates()
df_no_dup
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df["weather_condition"] = le.fit_transform(df["weather_condition"])

print(df.head())
```

```python
df.to_csv("cleaned_rainfall.csv",index=False)
```
