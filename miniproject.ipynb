Mini Project:

Create a Student Report Card Program using:

1.List → store student names

2.Set → store subjects

3.Tuple → store student information (name, age)

4.Dictionary → store subject-wise marks

Requirements:

1.Ask the user to enter:

*Student name

*Age

*Number of subjects

2.For each subject:

*Enter subject name

*Enter marks

3.Store:

*Name in a list

*Subjects in a set

*Student info in a tuple

*Marks in a dictionary

4.Print a report card showing:

*Name

*Age

*Subjects

*Marks

*Total marks

*Average marks

```python
std_names =[]
subjects=set()
marks_dict={}
name=input("Enter student name: ")
age=int(input("Enter student age: "))
num_subjects=int(input("Enter number of subjects: "))
std_names.append(name)
std_info=(name,age)
for i in range(num_subjects):
    sub=input(f"Enter subject {i+1} name:")
    marks=int(input(f"Enter marks in {sub}:"))
    subjects.add(sub)
    marks_dict[sub]=marks
total_marks=sum(marks_dict.values())
avg_marks=total_marks/num_subjects
print("Report Card:")
print("Name:",name)
print("Age:",age)
print("Subjects:",subjects)
total_marks = sum(marks_dict.values())
average_marks = total_marks / num_subjects
print("\nSubject-wise Marks:")
for sub,marks in marks_dict.items():
    print(f"{sub}:{marks}")
print("\nTotal Marks:",total_marks)
print("\nAverage Marks:",avg_marks)
```
