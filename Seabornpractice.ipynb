```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris)
plt.title('Iris ScatterPlot')
plt.show()
```

```python
sns.barplot(x='species',y='sepal_length',data=iris)
plt.title('Iris BarPlot')
plt.show()
```

```python
sns.histplot(x='sepal_length',hue='species',data=iris)
plt.title('Iris Histogram')
plt.show()
```

```python
sns.boxplot(x='species',y='sepal_length',data=iris)
plt.title('Iris BoxPlot')
plt.show()
```

```python
fig=figsize=(12,8)
sns.violinplot(x='species',y='sepal_length',data=iris)
plt.title('Iris ViolinPlot')
plt.show()
```

```python
sns.pairplot(iris,hue='species')
plt.title('Iris PairPlot')
plt.show()
```

```python
sns.heatmap(iris.corr(numeric_only=True),annot=True)
plt.title('Iris Heatmap')
plt.show()
```
